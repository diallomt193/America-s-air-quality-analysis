```python
#Import Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Save path to data set in a variable
data_file = "Resources/data.csv"
```

```python
#Read csv file

data_df = pd.read_csv(data_file)
data_df
```

```python
#Determine the average of pollutant released from 2010 to 2020 and add it in a new column named "Average"
#This average is calculated across each row
data_df["Average"] = data_df[["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean(axis = 1)
data_df
```

```python
#Let's also calculate the average of pollutant released from 2010 to 2020 across each column
Average = data_df[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean()
Average
```

```python
#Do a .loc to get rid of the years that we don't need and create a new dataframe containing only the data we want.
#In this case we are looking at years 2010 through 2020
data_df_copy = data_df.loc[:,["CBSA", "Core Based Statistical Area", "Pollutant", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "Average"]]
data_df_copy
```

```python
# Check the names of all the columns and see if there are any rows with missing data
data_df_copy.count()
```

```python
# Create a new DataFrame that looks into a specific pollutant SO2
SO2_pollutant_df = data_df_copy.loc[data_df_copy["Pollutant"] == "SO2"]
SO2_pollutant_df.drop_duplicates(subset = "Core Based Statistical Area", keep="first", inplace=True)
SO2_pollutant_df.reset_index(inplace=True)
SO2_pollutant_df
```

```python
#Print the statarea to see the complete list of cities.
statarea = SO2_pollutant_df["Core Based Statistical Area"]
print(statarea)
```

```python
#Pick a city in the list above and plot the data to show the change in S02 over a 10 year period "2010 - 2020"
#In this case, we picked Akron,OH because it is the first city in the list
#Next, create a dataframe for Akron,OH to observe the change of SO2 pollutant from 2010 to 2020
Akron_data = {'Year': [2010,2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020],
        'SO2_Rate': [27.0,24.0,22.0,23.0,21.0,14.0, 8.0, 3.0, 6.0, 3.0, 4.0]
       }
Akron_data_df = pd.DataFrame(Akron_data,columns=['Year','SO2_Rate'])
Akron_data_df
```

This data frame shows a decrease in SO2 release over the years, but the changes will be clearer in a graph (see below).

```python
#Plot the graph to observe the change
plt.plot(Akron_data_df['Year'], Akron_data_df['SO2_Rate'], color='red', marker='o')
plt.title('SO2 Rate Vs Year for Akron,OH', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('SO2 Rate', fontsize=14)
plt.savefig("Images/SO2 Rate Vs Year for Akron,OH.png")
plt.grid(True)
plt.show()
```

This graph shows that the release of SO2 has been decreasing over the years in Akron,OH

```python
#let's write a function that can plot a graph for each city in the list showing the change in NO2 from 2011 to 2020
#This function will plot a different graph every time we call it using the index number from the statearea.
def plot_graphs(i):
    Year = [2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020]
    city_data_df = pd.DataFrame(Akron_data,columns=['Year','SO2_Rate'])
    plt.plot(Year, SO2_pollutant_df.iloc[i, 6:16], color='red', marker='o')
    plt.title('SO2 Rate Vs Year for ' + SO2_pollutant_df.iloc[i, 3], fontsize=14)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('SO2 Rate', fontsize=14)
    plt.savefig("Images/SO2 Rate Vs Year for " + SO2_pollutant_df.iloc[i, 3]+".png")
    plt.grid(True)
    plt.show()
    
plot_graphs(0)
```

```python
#Now, let's plot the data for all cities to observe the change in SO2 over a five-year period "2016 - 2020"
#In this we decided to show the change for only 5 years because showing the change for 10 will make the graph hard to read.
plt.figure(figsize = (20, 7))
plt.plot(SO2_pollutant_df['Core Based Statistical Area'], SO2_pollutant_df['2016'], color='red', marker='o', label="2016")
plt.plot(SO2_pollutant_df['Core Based Statistical Area'], SO2_pollutant_df['2017'], color='blue', marker='o', label="2017")
plt.plot(SO2_pollutant_df['Core Based Statistical Area'], SO2_pollutant_df['2018'], color='green', marker='o', label="2018")
plt.plot(SO2_pollutant_df['Core Based Statistical Area'], SO2_pollutant_df['2019'], color='magenta', marker='o', label="2019")
plt.plot(SO2_pollutant_df['Core Based Statistical Area'], SO2_pollutant_df['2020'], color='yellow', marker='o', label="2020")
plt.xticks(rotation = 90)
plt.title('SO2 Rate Vs Year for different cities across the US', fontsize=14)
plt.xlabel('Core Based Statistical Area', fontsize=14)
plt.ylabel('SO2', fontsize=14)
plt.savefig("Images/SO2 Rate Vs Year for different cities across the US.png")
plt.grid(True)
plt.legend()
plt.show()
```

According to the graph, the city of Payson, AZ has the hightest level of SO2 release between the years 2016 and 2019, but then shows a significant decrease in 2020.
