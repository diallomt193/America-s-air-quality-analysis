```python
#Import Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Save path to data set in a variable
data_file = "Resources/data.csv"
```

```python
#Read csv file

data_df = pd.read_csv(data_file)
data_df
```

```python
#Determine the average of pollutant released from 2011 to 2020 and add it in a new column named "Average"
#This average is calculated across each row
data_df["Average"] = data_df[["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean(axis = 1)
data_df
```

```python
#Let's also calculate the average of pollutant released from 2010 to 2020 across each column
Average = data_df[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean()
Average
```

```python
#Do a .loc to get rid of the years that we don't need and create a new dataframe containing only the data we want.
#In this case we are looking at years 2011 through 2020
data_df_copy = data_df.loc[:,["CBSA", "Trend Statistic", "Core Based Statistical Area", "Pollutant", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "Average"]]
data_df_copy
```

```python
# Check the names of all the columns and see if there are any rows with missing data
data_df_copy.count()
```

```python
# Create a new DataFrame that looks into a specific pollutant NO2
NO2_pollutant_df = data_df_copy.loc[data_df_copy["Pollutant"] == "NO2"]
NO2_pollutant_df.drop_duplicates(subset = "Core Based Statistical Area", keep="first", inplace=True)
NO2_pollutant_df.reset_index(inplace=True)
NO2_pollutant_df
```

```python
#Print the statarea to see the complete list of cities.
statarea = NO2_pollutant_df["Core Based Statistical Area"]
print(statarea)
```

```python
#Pick a city in the list above and plot the data to show the change in NO2 over a 10 year period "2011 - 2020"
#In this case, we picked Albuquerque,NM because it is the first city in the list
#Next, create a dataframe for Albuquerque,NM to observe the change of O3 pollutant from 2011 to 2020
Albuquerque_data = {'Year': [2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020],
        'NO2_Rate': [13.0, 14.0, 12.0, 12.0, 11.0, 10.0, 10.0, 10.0, 9.0, 8.0]
       }
Albuquerque_data_df = pd.DataFrame(Albuquerque_data,columns=['Year','NO2_Rate'])
Albuquerque_data_df
```

This data frame shows a decrease in NO2 release over the years, but the changes will be clearer in a graph (see below).

```python
plt.plot(Albuquerque_data_df['Year'], Albuquerque_data_df['NO2_Rate'], color='red', marker='o')
plt.title('NO2 Rate Vs Year for Albuquerque, NM', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('NO2 Rate', fontsize=14)
plt.savefig("Images/NO2 Rate Vs Year for Albuquerque, NM.png")
plt.grid(True)
plt.show()
```

This graph shows that the release of SO2 has been decreasing over the years in Albuquerque,NM

```python
#let's write a function that can plot a graph for each city in the list showing the change in NO2 from 2011 to 2020
#This function will plot a different graph every time we call it using the index number from the statearea.
def plot_graphs(i):
    Year = [2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020]
    city_data_df = pd.DataFrame(Albuquerque_data,columns=['Year','NO2_Rate'])
    plt.plot(Year, NO2_pollutant_df.iloc[i, 5:15], color='red', marker='o')
    plt.title('NO2 Rate Vs Year for ' + NO2_pollutant_df.iloc[i, 3], fontsize=14)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('NO2 Rate', fontsize=14)
    plt.savefig("Images/NO2 Rate Vs Year for " + NO2_pollutant_df.iloc[i, 3]+".png")
    plt.grid(True)
    plt.show()
    
plot_graphs(17)
```

```python
#Let's sort the data from highest to lowest average rate of NO2 released between 2011 to 2020 and plot those values to get a cleaner graph
NO2_change = NO2_pollutant_df.sort_values(["Average"], ascending=False)
NO2_change
```

```python
#Plot 2 graphs and compare them
#First graph will show the average rate of NO2 released between 2011 and 2020, and the second graph will have the NO2 rates for the 2020
plt.figure(figsize = (40, 10))
plt.plot(NO2_change['Core Based Statistical Area'], NO2_change['Average'], color='red', marker='o', label = '10-year Average')
plt.plot(NO2_change['Core Based Statistical Area'], NO2_change['2020'], color='blue', marker='o', label = '2020')
plt.xticks(rotation = 90)
plt.title('NO2 Rate Vs Year for different cities across the US (Average vs 2020)', fontsize=14)
plt.xlabel('Core Based Statistical Area', fontsize=14)
plt.ylabel('NO2', fontsize=14)
plt.grid(True)
plt.legend()
plt.xlim(-0.5,len(NO2_change['Core Based Statistical Area']))
plt.savefig("Images/NO2 Rate Vs Year for different cities across the US (Average vs 2020).png")
plt.tight_layout()
plt.show()
```

According to this graph at least 4 cities in the year 2020 had a NO2 rate higher than their average.

```python
#Now, let's plot the data for all cities to observe the change in NO2 over a five-year period "2016 - 2020"
#In this we decided to show the change for only 5 years because showing the change for 10 will make the graph hard to read.
plt.figure(figsize = (20, 5))
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2016'], color='red', marker='o', label="2016")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2017'], color='blue', marker='o', label="2017")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2018'], color='green', marker='o', label="2018")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2019'], color='magenta', marker='o', label="2019")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2020'], color='yellow', marker='o', label="2020")
plt.xticks(rotation = 90)
plt.title('NO2 Rate Vs Year for different cities across the US', fontsize=14)
plt.xlabel('Core Based Statistical Area', fontsize=14)
plt.ylabel('NO2', fontsize=14)
plt.savefig("Images/NO2 Rate Vs Year for different cities across the US.png")
plt.grid(True)
plt.legend()
plt.show()
```
