```python
#Import Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Save path to data set in a variable
data_file = "Resources/data.csv"
```

```python
#Read csv file

data_df = pd.read_csv(data_file)
data_df
```

```python
#Determine the average of pollutant released through the years
#First add an empty column to hold the data
data_df["Average"] = data_df[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean(axis = 1)
data_df
```

```python
#Get rid of the years we don't need (2000 -2009)
data_df_copy = data_df.loc[:,["CBSA", "Trend Statistic", "Core Based Statistical Area", "Pollutant", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "Average"]]
data_df_copy
```

```python
# Check the names of all the columns and see if there are any rows with missing data
data_df_copy.count()
```

```python
# Create a new DataFrame that looks into a specific pollutant
NO2_pollutant_df = data_df_copy.loc[data_df_copy["Pollutant"] == "NO2"]
NO2_pollutant_df.drop_duplicates(subset = "Core Based Statistical Area", keep="first", inplace=True)
NO2_pollutant_df.reset_index(inplace=True)
NO2_pollutant_df
```

```python
statarea = NO2_pollutant_df["Core Based Statistical Area"]
print(statarea)
```

```python
Albuquerque_data = {'Year': [2010,2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020],
        'NO2_Rate': [12.0, 13.0, 14.0, 12.0, 12.0, 11.0, 10.0, 10.0, 10.0, 9.0, 8.0]
       }
Albuquerque_data_df = pd.DataFrame(Albuquerque_data,columns=['Year','NO2_Rate'])
Albuquerque_data_df
```

```python
plt.plot(Albuquerque_data_df['Year'], Albuquerque_data_df['NO2_Rate'], color='red', marker='o')
plt.title('NO2 Rate Vs Year for Albuquerque, NM', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('NO2 Rate', fontsize=14)
plt.grid(True)
plt.show()
```

```python
#let's plot the data
plt.figure(figsize = (20, 5))
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2016'], color='red', marker='o', label="2016")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2017'], color='blue', marker='o', label="2017")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2018'], color='green', marker='o', label="2018")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2019'], color='magenta', marker='o', label="2019")
plt.plot(NO2_pollutant_df['Core Based Statistical Area'], NO2_pollutant_df['2020'], color='yellow', marker='o', label="2020")
plt.xticks(rotation = 90)
plt.title('NO2 Rate Vs Year for different cities across the US', fontsize=14)
plt.xlabel('Core Based Statistical Area', fontsize=14)
plt.ylabel('NO2', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
