```python
#Import Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Save path to data set in a variable
data_file = "Resources/data.csv"
```

```python
#Read csv file

data_df = pd.read_csv(data_file)
data_df
```

```python
#Determine the average of pollutant released through the years
#First add an empty column to hold the data
data_df["Average"] = data_df[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean(axis = 1)
data_df
```

```python
#Get rid of the years we don't need (2000 -2009)
data_df_copy = data_df.loc[:,["CBSA", "Trend Statistic", "Core Based Statistical Area", "Pollutant", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "Average"]]
data_df_copy
```

```python
# Check the names of all the columns and see if there are any rows with missing data
data_df_copy.count()
```

```python
# Create a new DataFrame that looks into a specific pollutant
O3_pollutant_df = data_df_copy.loc[data_df_copy["Pollutant"] == "O3"]
O3_pollutant_df.drop_duplicates(subset = "Core Based Statistical Area", keep="first", inplace=True)
O3_pollutant_df.reset_index(inplace=True)
O3_pollutant_df
```

```python
statarea = O3_pollutant_df["Core Based Statistical Area"]
print(statarea)
```

```python
Adrian_data = {'Year': [2010,2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020],
        'O3_Rate': [0.066, 0.076, 0.087, 0.064, 0.068, 0.065, 0.069, 0.066, 0.071, 0.059, 0.067]
       }
Adrian_data_df = pd.DataFrame(Adrian_data,columns=['Year','O3_Rate'])
Adrian_data_df
```

```python
plt.plot(Adrian_data_df['Year'], Adrian_data_df['O3_Rate'], color='red', marker='o')
plt.title('O3 Rate Vs Year for Adrian,MI', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('O3 Rate', fontsize=14)
plt.grid(True)
plt.show()
```

```python
#let's plot the data
plt.figure(figsize = (30, 20))
plt.plot(O3_pollutant_df['Core Based Statistical Area'], O3_pollutant_df['2016'], color='red', marker='o', label="2016")
plt.plot(O3_pollutant_df['Core Based Statistical Area'], O3_pollutant_df['2017'], color='blue', marker='o', label="2017")
plt.plot(O3_pollutant_df['Core Based Statistical Area'], O3_pollutant_df['2018'], color='green', marker='o', label="2018")
plt.plot(O3_pollutant_df['Core Based Statistical Area'], O3_pollutant_df['2019'], color='magenta', marker='o', label="2019")
plt.plot(O3_pollutant_df['Core Based Statistical Area'], O3_pollutant_df['2020'], color='yellow', marker='o', label="2020")
plt.xticks(rotation = 90)
plt.title('O3 Rate Vs Year for different cities across the US', fontsize=14)
plt.xlabel('Core Based Statistical Area', fontsize=14)
plt.ylabel('SO2', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
