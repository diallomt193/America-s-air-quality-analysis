```python
# Save path to data set in a variable
data_file = "Resources/data.csv"
```

```python
#Read csv file

data_df = pd.read_csv(data_file)
data_df
```

```python
data_df_copy = data_df.loc[:,["CBSA", "Trend Statistic", "Core Based Statistical Area", "Pollutant", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020",]]
data_df_copy
```

```python
# Create a new DataFrame that looks into a specific neighborhood
O3_pollutant_df = data_df_copy.loc[data_df_copy["Pollutant"] == "O3"]
O3_pollutant_df
```

```python
Adrian_data = {'Year': [2010,2011,2012,2013,2014,2015, 2016, 2017, 2018, 2019, 2020],
        'O3_Rate': [0.066, 0.076, 0.087, 0.064, 0.068, 0.065, 0.069, 0.066, 0.071, 0.059, 0.067]
       }
Adrian_data_df = pd.DataFrame(Adrian_data,columns=['Year','O3_Rate'])
Adrian_data_df
```

```python
plt.plot(Adrian_data_df['Year'], Adrian_data_df['O3_Rate'], color='red', marker='o')
plt.title('O3 Rate Vs Year for Adrian,MI', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('O3 Rate', fontsize=14)
plt.grid(True)
plt.show()
```
